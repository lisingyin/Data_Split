```python
def replacement(x):
    rest_type = pd.read_excel("Restaurant_TYPE.xlsx")
    for y in range(0,33):
        if (x == rest_type.iloc[y,1]):
            return rest_type.iloc[y,0]
        else:
            continue
            
```

```python
abc = restaurant_type["T1"].apply(replacement)
abc = abc.append(restaurant_type["T2"].apply(replacement))
abc = abc.append(restaurant_type["T3"].apply(replacement))
abc = abc.append(restaurant_type["T4"].apply(replacement))
abc = abc.append(restaurant_type["T5"].apply(replacement))
abc.dropna(inplace=True)
FYP_type["ID"] = abc.index
FYP_type["Type_ID"] = abc.values
```
